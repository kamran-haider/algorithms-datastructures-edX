```python
def pairwise_product_fast(a):
    result = 0
    # find first max
    index_1 = 0
    for i in range(len(a)):
        if a[i] > a[index_1]:
            index_1 = i
    # if first element is max, look for next max from index 1 onwards 
    if index_1 == 0:
        index_2 = 1
    else:
        index_2 = 0
    #print(index_1, index_2)
    for i in range(len(a)):
        # if current value is greater than current index_2 value
        # if current value is not equal to index_1 value
        if a[i] > a[index_2] and i != index_1:
            index_2 = i
    return a[index_1] * a[index_2]

def pairwise_product_naive(a):
    result = 0
    n = len(a)
    for i in range(n):
        for j in range(i+1, n):
            if a[i]*a[j] > result:
                result = a[i]*a[j]
    return result
print(pairwise_product_fast([0, 1, 0, 0]))
print(pairwise_product_naive([0, 1, 0, 0]))
```

```python
def stress_test(N, M):
    """
    Stress testing of the maximum pairwise product algorithms.
    
    Parameters
    ----------
    N : int
        Maximum length of array, in each iteration a random number
        between 2 and N is drawn to set the array length.
    M : int
        Maximum value of the numbers in array, each array contains numbers
        between 0 and M.
    
    Return
    ------
    result_1, result_2 : int, int
        Naive and fast answers from the mismatch
    """
    passed = True
    while passed:
        n = np.random.randint(2, N)
        a = np.random.randint(0, M, n)
        result_1 = pairwise_product_naive(a)
        result_2 = pairwise_product_fast(a)
        print(a)
        if result_1 == result_2:
            print("OK")
        else:
            print("Wrong Answer")
            passed = False
    return result_1, result_2
```
